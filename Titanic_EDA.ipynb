```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df.drop("Cabin", axis=1, inplace=True)

df.isnull().sum()
```

```python
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(df["Fare"], bins=30, kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

# Conclusion

The Titanic dataset was cleaned by handling missing values and removing the Cabin column due to excessive missing data.

Key findings:

- Female passengers had higher survival rates than males.
- First-class passengers survived more frequently than third-class passengers.
- Higher ticket fares were associated with better survival chances.
- Most passengers were between 20 and 40 years old.
- Passenger class and gender strongly influenced survival.

This analysis demonstrates how data cleaning and EDA help uncover patterns and relationships in a dataset.
